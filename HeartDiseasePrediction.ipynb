```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training(Logistic Regression)

```python
model = LogisticRegression()
```

```python
# training the LogisticRegression model with Training data
model.fit(X_train, Y_train)
```

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```

Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset into a DataFrame
heart_data = pd.read_csv(r'C:\Users\hp\Desktop\heart_disease_data.csv')

# Check the first few rows to ensure data is loaded correctly
print(heart_data.head())

# Print the column names to confirm structure
print("Column names:", heart_data.columns.tolist())

# 1. Histogram for Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(heart_data['age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 2. Pie chart for Heart Disease distribution based on 'target' column
target_counts = heart_data['target'].value_counts()

# Create labels for the pie chart
labels = ['No Heart Disease (0)', 'Has Heart Disease (1)']

# Draw pie chart
plt.figure(figsize=(8, 8))
plt.pie(target_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Heart Disease Distribution')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```
